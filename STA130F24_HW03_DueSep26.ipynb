````python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import seaborn as sns

# Load the penguins dataset
penguins = sns.load_dataset('penguins')

# Remove rows with missing values for the purpose of this example
penguins = penguins.dropna(subset=['flipper_length_mm'])

# Define a function to create histograms with the required annotations
def plot_species_histograms(df, species):
    fig = go.Figure()

    for sp in species:
        # Filter the data for the species
        sp_data = df[df['species'] == sp]['flipper_length_mm']
        
        # Calculate statistics
        mean = sp_data.mean()
        median = sp_data.median()
        min_val = sp_data.min()
        max_val = sp_data.max()
        std = sp_data.std()
        q1 = sp_data.quantile(0.25)
        q3 = sp_data.quantile(0.75)
        
        # Add histogram for the species
        fig.add_trace(go.Histogram(x=sp_data, name=f'{sp} Flipper Length', opacity=0.7))
        
        # Add vertical lines for mean and median
        fig.add_vline(x=mean, line_dash="dash", line_color="green", annotation_text=f"{sp} Mean", annotation_position="top left")
        fig.add_vline(x=median, line_dash="dot", line_color="blue", annotation_text=f"{sp} Median", annotation_position="top left")
        
        # Add vertical rectangles for range, IQR, and 2 standard deviations
        fig.add_vrect(x0=min_val, x1=max_val, line_width=0, fillcolor="gray", opacity=0.1, annotation_text=f"{sp} Range", annotation_position="bottom left")
        fig.add_vrect(x0=q1, x1=q3, line_width=0, fillcolor="orange", opacity=0.2, annotation_text=f"{sp} IQR", annotation_position="bottom left")
        fig.add_vrect(x0=mean - 2*std, x1=mean + 2*std, line_width=0, fillcolor="purple", opacity=0.2, annotation_text=f"{sp} 2", annotation_position="bottom left")
    
    # Update layout
    fig.update_layout(
        title="Flipper Length Distribution with Mean, Median, and Ranges",
        barmode='overlay',
        xaxis_title="Flipper Length (mm)",
        yaxis_title="Count",
        template="plotly_white"
    )

    fig.show(renderer="png")

# Call the function with penguins data
species = penguins['species'].unique()
plot_species_histograms(penguins, species)
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the penguins dataset
penguins = sns.load_dataset('penguins')

# Remove rows with missing values
penguins = penguins.dropna(subset=['flipper_length_mm'])

# Get the list of unique species
species = penguins['species'].unique()

# Set up the figure and axes (one row, three columns for the three species)
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Define a function to plot KDE with vertical lines and shaded regions
def plot_kde_with_annotations(ax, data, species_name):
    # Calculate statistics
    mean = data.mean()
    median = data.median()
    min_val = data.min()
    max_val = data.max()
    std = data.std()
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    
    # Plot KDE
    sns.kdeplot(data, ax=ax, fill=True, color="skyblue", label=f"{species_name} KDE")
    
    # Add vertical lines for mean and median
    ax.axvline(mean, color='green', linestyle='--', label='Mean', lw=2)
    ax.axvline(median, color='blue', linestyle='-.', label='Median', lw=2)
    
    # Add shaded regions for range, IQR, and 2
    ax.axvspan(min_val, max_val, color='gray', alpha=0.1, label='Range')
    ax.axvspan(q1, q3, color='orange', alpha=0.2, label='IQR')
    ax.axvspan(mean - 2*std, mean + 2*std, color='purple', alpha=0.2, label='2')
    
    # Set title and labels
    ax.set_title(f"{species_name} Flipper Length KDE", fontsize=14)
    ax.set_xlabel("Flipper Length (mm)", fontsize=12)
    ax.set_ylabel("Density", fontsize=12)

# Loop over species to create the KDE plots
for i, sp in enumerate(species):
    species_data = penguins[penguins['species'] == sp]['flipper_length_mm']
    plot_kde_with_annotations(axes[i], species_data, sp)

# Add a legend to the first plot
axes[0].legend(loc='upper right', fontsize=10)

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
````

Heres a summary of our discussion:

1. **Data Visualization Methods**:
   - We discussed three methods for visualizing data distribution: **boxplots**, **histograms**, and **KDE (Kernel Density Estimation)** plots.
   
2. **Boxplots**:
   - **Pros**: Provide summary statistics (minimum, Q1, median, Q3, maximum), highlight outliers, and allow for easy comparison between groups.
   - **Cons**: Lose detail about the data distribution within quartiles, may mislead if the data is not symmetrically distributed, and can be less effective with small samples.

3. **Histograms**:
   - **Pros**: Clearly visualize the distribution of data, show where values are concentrated, are flexible with bin width, and are easy to interpret.
   - **Cons**: Sensitive to bin width choice, can obscure individual data points, and can be challenging for comparing multiple distributions.

4. **KDE (Kernel Density Estimation)**:
   - **Pros**: Provides a smooth representation of data distribution, is flexible with bandwidth adjustments, and facilitates easy comparison of distributions.
   - **Cons**: Sensitive to bandwidth selection, may be less intuitive for some audiences, and can be computationally intensive for large datasets.

mean is the average which is not independent, median is the typical value which is independent.

When it's right skewed, mean is greater than the median. Because high values (outliers) on the right pulls the mean to the right, while the median is affected by extreme values.

When it's left skewed, mean is smaller than the median. Because high values (outliers) on the left pulls the mean to the left, while the median is affected by extreme values.

````python
import plotly.express as px
import pandas as pd
import numpy as np
from scipy import stats

# Generate a right-skewed sample (Gamma distribution)
sample1 = stats.gamma(a=2, scale=2).rvs(size=1000)

# Calculate mean and median for the right-skewed sample
mean1 = sample1.mean()
median1 = np.quantile(sample1, [0.5])

# Generate a left-skewed sample (Negative of Gamma distribution)
sample2 = -stats.gamma(a=2, scale=2).rvs(size=1000)

# Calculate mean and median for the left-skewed sample
mean2 = sample2.mean()
median2 = np.quantile(sample2, [0.5])

# Create histograms
fig1 = px.histogram(pd.DataFrame({'data': sample1}), x="data", title='Right-Skewed Distribution (Gamma)')
fig2 = px.histogram(pd.DataFrame({'data': sample2}), x="data", title='Left-Skewed Distribution (-Gamma)')

# Add mean and median lines to the plots
fig1.add_vline(x=mean1, line_color='red', line_width=2, annotation_text='Mean', annotation_position='top left')
fig1.add_vline(x=median1, line_color='blue', line_width=2, annotation_text='Median', annotation_position='top left')

fig2.add_vline(x=mean2, line_color='red', line_width=2, annotation_text='Mean', annotation_position='top left')
fig2.add_vline(x=median2, line_color='blue', line_width=2, annotation_text='Median', annotation_position='top left')

# Display the figures
fig1.show(renderer="png")
fig2.show(renderer="png")

# Print mean and median values for both samples
print("Right-Skewed Sample:")
print(f"Mean: {mean1}, Median: {median1}")

print("\nLeft-Skewed Sample:")
print(f"Mean: {mean2}, Median: {median2}")
````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
url = 'https://raw.githubusercontent.com/canzafar/World-Happiness-Report/main/2023.csv'
data = pd.read_csv(url)

# Display the first few rows of the dataset
print("First few rows of the dataset:\n", data.head())

# Check the columns to see the actual names
print("\nColumns in the dataset:\n", data.columns.tolist())

# Check for missing values
print("\nMissing values in each column:\n", data.isnull().sum())

# Summary statistics
summary_stats = data.describe()
print("\nSummary Statistics:\n", summary_stats)

# Ensure that the expected columns exist before visualizing
if 'Happiness Score' in data.columns and 'GDP per Capita' in data.columns:
    # Visualizing the distribution of Happiness Scores
    plt.figure(figsize=(12, 6))
    sns.histplot(data['Happiness Score'], kde=True, bins=30)
    plt.title('Distribution of Happiness Scores')
    plt.xlabel('Happiness Score')
    plt.ylabel('Frequency')
    plt.show()

    # Visualizing Happiness Score vs. GDP per Capita
    plt.figure(figsize=(12, 6))
    sns.scatterplot(data=data, x='GDP per Capita', y='Happiness Score', hue='Region', palette='deep')
    plt.title('Happiness Score vs. GDP per Capita')
    plt.xlabel('GDP per Capita')
    plt.ylabel('Happiness Score')
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
    plt.show()

    # Correlation heatmap
    plt.figure(figsize=(10, 8))
    correlation = data.corr()
    sns.heatmap(correlation, annot=True, fmt=".2f", cmap='coolwarm')
    plt.title('Correlation Heatmap')
    plt.show()
else:
    print("One or more required columns are missing in the dataset.")
````

````python
import pandas as pd
import plotly.express as px

# Load the baby names dataset
bn = pd.read_csv('https://raw.githubusercontent.com/hadley/data-baby-names/master/baby-names.csv')

# Make identical boy and girl names distinct
bn['name'] = bn['name'] + " " + bn['sex']

# Calculate rank and sort values
bn['rank'] = bn.groupby('year')['percent'].rank(ascending=False)
bn = bn.sort_values(['name', 'year'])

# Create percent change from the previous year
bn['percent change'] = bn['percent'].diff()
new_name = [True] + list(bn.name[:-1].values != bn.name[1:].values)
bn.loc[new_name, 'percent change'] = bn.loc[new_name, 'percent'] 

# Sort and restrict to common names
bn = bn.sort_values('year')
bn = bn[bn.percent > 0.001]  # Keep only common names

# Create the scatter plot with the specified parameters
fig = px.scatter(bn, 
                 x="percent change", 
                 y="rank", 
                 animation_frame="year", 
                 animation_group="name",
                 size="percent", 
                 color="sex", 
                 hover_name="name",
                 size_max=50, 
                 range_x=[-0.005, 0.005])  # range_y parameter removed

# Reverse the y-axis to have rank 1 on top
fig.update_yaxes(autorange='reversed')

# Show the figure
fig.show(renderer="png")  # USE `fig.show(renderer="png")` FOR ALL GitHub and MarkUs SUBMISSIONS
````
