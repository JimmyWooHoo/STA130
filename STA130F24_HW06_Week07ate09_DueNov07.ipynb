The outcome Y is based on a starting point (Œ≤0), adjusted by the predictor‚Äôs influence (ùõΩ1√óùëã), with a bit of error(œµ) thrown in. In Simple Linear Regression, the predictor variable and coefficients create a line for the outcome, and by adding normally distributed error, each outcome value follows a normal distribution around this line.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set the parameters for the Simple Linear Regression model
beta_0 = 5  # intercept
beta_1 = 2  # slope
sigma = 3   # standard deviation of the error term

# Generate sample predictor data (X)
np.random.seed(0)  # for reproducibility
X = np.linspace(0, 10, 50)

# Calculate the true outcome (Y) based on the linear equation without error
Y_true = beta_0 + beta_1 * X

# Add normally distributed error to simulate actual observed Y values
error = np.random.normal(0, sigma, size=X.shape)
Y_observed = Y_true + error

# Plot the results
plt.figure(figsize=(10, 6))

# Plot the true regression line (without error)
plt.plot(X, Y_true, color="blue", label="True Regression Line (no error)")

# Plot the observed Y values with error (scatter plot)
plt.scatter(X, Y_observed, color="orange", label="Observed Y (with error)")

# Show a normal distribution for Y at a specific X value (e.g., X=5)
x_value = 5
mean_y_at_x = beta_0 + beta_1 * x_value
y_values = np.linspace(mean_y_at_x - 3*sigma, mean_y_at_x + 3*sigma, 100)
pdf_y = norm.pdf(y_values, mean_y_at_x, sigma)
plt.plot([x_value]*100, y_values, 'r--', label=f"Normal Distribution of Y | X={x_value}")
plt.fill_betweenx(y_values, x_value - pdf_y*0.5, x_value + pdf_y*0.5, color="gray", alpha=0.2)

# Labels and legend
plt.xlabel("Predictor (X)")
plt.ylabel("Outcome (Y)")
plt.title("Simple Linear Regression with Normally Distributed Errors")
plt.legend()
plt.show()
````

create and visualize a fitted Simple Linear Regression model:

- We generated a simulated dataset with predictor \( X \) and outcome \( Y \) (including normally distributed error).
- Using `statsmodels`, we fitted a regression model with \( Y \) as the dependent variable and \( X \) as the predictor.
- The summary output shows model details, including estimates of the intercept and slope coefficients, their significance, and other fit statistics.
- The scatter plot (orange points) shows the simulated data, and the fitted regression line (blue line) represents the model‚Äôs best-fit line based on this data.

````python
import pandas as pd
import statsmodels.formula.api as smf

# Generate a simulated dataset based on the theoretical Simple Linear Regression model
np.random.seed(0)  # for reproducibility
X_simulated = np.linspace(0, 10, 50)
error_simulated = np.random.normal(0, sigma, size=X_simulated.shape)
Y_simulated = beta_0 + beta_1 * X_simulated + error_simulated

# Create a pandas DataFrame to hold the data
data = pd.DataFrame({'X': X_simulated, 'Y': Y_simulated})

# Fit a Simple Linear Regression model using statsmodels
model = smf.ols('Y ~ X', data=data).fit()

# Print model summary
print(model.summary())

# Plot the observed data and the fitted regression line
plt.figure(figsize=(10, 6))

# Scatter plot of the simulated data
plt.scatter(data['X'], data['Y'], color="orange", label="Simulated Observed Data")

# Plot the fitted regression line
plt.plot(data['X'], model.fittedvalues, color="blue", label="Fitted Regression Line")

# Labels and legend
plt.xlabel("Predictor (X)")
plt.ylabel("Outcome (Y)")
plt.title("Fitted Simple Linear Regression Model")
plt.legend()
plt.show()
````

>

|p-value|Evidence|
|-|-|
|$$p > 0.1$$|No evidence against the null hypothesis|
|$$0.1 \ge p > 0.05$$|Weak evidence against the null hypothesis|
|$$0.05 \ge p > 0.01$$|Moderate evidence against the null hypothesis|
|$$0.01 \ge p > 0.001$$|Strong evidence against the null hypothesis|
|$$0.001 \ge p$$|Very strong evidence against the null hypothesis|

  - If the p-value is large, we fail to reject \( H_0 \), meaning no significant linear relationship.

### Interpretation for the Old Faithful Geyser:
- A small p-value means waiting time significantly affects eruption duration.

# 1. Bootstrapping for Fitted Slope Coefficients:

````python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Number of bootstrap samples
n_bootstraps = 1000

# Collecting bootstrapped slope coefficients
bootstrap_slopes = []

for _ in range(n_bootstraps):
    # Create a bootstrap sample
    bootstrap_sample = old_faithful[long_wait].sample(n=160, replace=True)
    
    # Fit a Simple Linear Regression model
    X = sm.add_constant(bootstrap_sample['waiting'])  # Adding intercept
    y = bootstrap_sample['duration']
    
    model = sm.OLS(y, X).fit()
    
    # Collect the slope coefficient
    bootstrap_slopes.append(model.params['waiting'])

# Visualize the bootstrapped sampling distribution
plt.hist(bootstrap_slopes, bins=30, edgecolor='k', alpha=0.7)
plt.title('Bootstrapped Sampling Distribution of Fitted Slope Coefficients')
plt.xlabel('Slope Coefficients')
plt.ylabel('Frequency')
plt.show()
````

# 2. Simulate Samples Under the Null Hypothesis (No Linear Association):

````python
# Parameters under the null hypothesis
beta_0 = 1.65
beta_1 = 0
sigma = 0.37

# Simulate samples and collect fitted slope coefficients
simulated_slopes = []
n_simulations = 1000

for _ in range(n_simulations):
    # Simulate Y values based on the null hypothesis
    X_simulated = old_faithful[long_wait]['waiting']
    noise = np.random.normal(0, sigma, size=len(X_simulated))
    Y_simulated = beta_0 + beta_1 * X_simulated + noise
    
    # Fit a model to the simulated data
    X_simulated_const = sm.add_constant(X_simulated)  # Adding intercept
    model_simulated = sm.OLS(Y_simulated, X_simulated_const).fit()
    
    # Collect the fitted slope coefficient
    simulated_slopes.append(model_simulated.params['waiting'])

# Visualize the sampling distribution of the simulated slopes
plt.hist(simulated_slopes, bins=30, edgecolor='k', alpha=0.7)
plt.title('Sampling Distribution of Fitted Slope Coefficients under the Null Hypothesis')
plt.xlabel('Slope Coefficients')
plt.ylabel('Frequency')
plt.show()
````

# 3. Bootstrapped Confidence Interval and p-value Comparison:

````python
# Bootstrapped Confidence Interval
lower_bound = np.percentile(bootstrap_slopes, 2.5)
upper_bound = np.percentile(bootstrap_slopes, 97.5)
print(f'Bootstrapped 95% CI for slope: ({lower_bound:.4f}, {upper_bound:.4f})')

# Check if 0 is contained in the confidence interval
if lower_bound < 0 < upper_bound:
    print("0 is contained within the 95% bootstrapped confidence interval.")
else:
    print("0 is not contained within the 95% bootstrapped confidence interval.")
````

````python
# Get the actual fitted slope from the full dataset
X_full = sm.add_constant(old_faithful[long_wait]['waiting'])
y_full = old_faithful[long_wait]['duration']
model_full = sm.OLS(y_full, X_full).fit()
actual_slope = model_full.params['waiting']

# Calculate the p-value based on the simulated slopes
simulated_p_value = np.mean(np.abs(np.array(simulated_slopes)) >= np.abs(actual_slope))
print(f'Simulated p-value: {simulated_p_value:.4f}')
````

### 13. The "short" and "long" wait times are not "before and after" measurements so there are not natural pairs on which to base differences on which to do a "one sample" (paired differences) *hypothesis test*; but, we can do "two sample" hypothesis testing using a *permuation test*, or create a 95% *bootstrap confidence interval* for the difference in means of the two populations. 

### (A) Do a permuation test $\;H_0: \mu_{\text{short}}=\mu_{\text{long}} \; \text{ no difference in duration between short and long groups}$ by "shuffling" the labels
### (B) Create a 95% bootstrap confidence interval  by repeatedly bootstrapping within each group and applying *np.quantile(bootstrapped_mean_differences, [0.025, 0.975])* to the collection of differences between the sample means.    
### (a) Explain how the sampling approaches work for the two simulations.
### (b) Compare and contrast these two methods with the *indicator variable* based model approach used in Question 11, explaining how they're similar and different.<br>
    
<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>

> _You'll need to create `for` loops for repeated (shuffling simulation) **permutation** and (subgroup) **bootstrapping**, where_
>
> - _"shuffling" for **permutation testing** is done like this `old_faithful.assign(kind_shuffled=old_faithful['kind'].sample(n=len(old_faithful), replace=False).values)#.groupby('kind').size()`; then, the **mean difference statistic** is then calculated using `.groupby('kind_shuffled')['duration'].mean().iloc[::-1].diff().values[1]` (so the **observed statistic** is `old_faithful.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]`_
> 
>
> - _"two sample" **bootstrapping** is done like this `old_faithful.groupby('kind').apply(lambda x: x.sample(n=len(x), replace=True)).reset_index(drop=True)#.groupby('kind').size()`; then, the **bootstrapped mean difference statistic** is then calculated using `.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]` (like the **observed statistic** except this is applied to the **bootstrapped** resampling of `old_faithful`)_
> ---
> 
> _Don't forget to ask for summaries of all your different ChatBot sessions and organize and paste these into your homework notebook (including link(s) to chat log histories if you're using ChatBot) But if you're using the STA130 custom NBLM ChatBot, you'll only be able to ask for summaries, of course!_
</details>

````python
# A
import numpy as np
import pandas as pd

# Define the observed statistic (difference in means between the short and long groups)
observed_statistic = old_faithful.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]

# Number of permutations (shuffles)
n_permutations = 10000
permutation_diffs = []

# Perform permutation test by shuffling the 'kind' labels
for _ in range(n_permutations):
    # Shuffle the "kind" column (short/long groups)
    shuffled_data = old_faithful.assign(kind_shuffled=old_faithful['kind'].sample(n=len(old_faithful), replace=False).values)
    
    # Calculate the difference in means between the shuffled groups
    perm_statistic = shuffled_data.groupby('kind_shuffled')['duration'].mean().iloc[::-1].diff().values[1]
    permutation_diffs.append(perm_statistic)

# Calculate p-value for the permutation test
permutation_p_value = np.mean(np.abs(permutation_diffs) >= np.abs(observed_statistic))

# Print the observed statistic and p-value
print(f"Observed Difference in Means: {observed_statistic}")
print(f"Permutation Test p-value: {permutation_p_value}")
````

````python
# B
import numpy as np
import pandas as pd

# Number of bootstrap iterations
n_bootstrap = 10000
bootstrap_diffs = []

# Perform bootstrapping for the short and long wait time groups
for _ in range(n_bootstrap):
    # Bootstrap resampling for the "short" and "long" groups
    bootstrapped_data = old_faithful.groupby('kind').apply(lambda x: x.sample(n=len(x), replace=True)).reset_index(drop=True)
    
    # Calculate the difference in means between the bootstrapped "short" and "long" groups
    boot_statistic = bootstrapped_data.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]
    bootstrap_diffs.append(boot_statistic)

# Calculate the 95% bootstrap confidence interval
ci_lower = np.percentile(bootstrap_diffs, 2.5)
ci_upper = np.percentile(bootstrap_diffs, 97.5)

# Print the bootstrap confidence interval
print(f"Bootstrap 95% Confidence Interval for the Difference in Means: ({ci_lower}, {ci_upper})")
````

(a) In both cases, the idea is to generate multiple samples from the observed data to understand the variability in the statistic (difference in means) and assess whether the observed difference is statistically significant. However, the permutation test focuses on testing the null hypothesis by comparing the observed statistic to the distribution of shuffled statistics, while bootstrapping focuses on estimating the variability of the statistic through repeated resampling within each group.

(b) The indicator variable model uses a direct parametric approach to compare group differences, while the permutation and bootstrap methods focus on resampling to estimate the difference without making strong assumptions about the data's distribution.

### Ethical Professionalism Considerations
    
The TUT and HW both addressed some of the assumptions used in **Simple Linear Regression**. The **p-values** provided by `statsmodels` via `smf.ols(...).fit()` depend on these assumptions, so if they are not (at least approximately) correct, the **p-values** (and any subsequent claims regarding the "evidience against" the **null hypothesis**) are not reliable. In light of this consideration, describe how you could diagnostically check the first three assumptions (given below) when using analyses based on **Simple Linear regression** model. From an Ethical and Professional perspective, do you think doing diagnostic checks on the assumptions of a **Simple Linear regression** model is something you can and should do whenever you're doing this kind of analysis? 
            
> The first three assumptions associated with the **Simple Linear regression** model are that
> 
> - the $\epsilon_i$ **errors** (sometimes referred to as the **noise**) are **normally distributed**
> - the $\epsilon_i$ **errors** are **homoscedastic** (so their distributional variance $\sigma^2$ does not change as a function of $x_i$)
> - the linear form is [at least reasonably approximately] "true" (in the sense that the above two remain [at least reasonably approximately] "true") so that then behavior of the $Y_i$ **outcomes** are represented/determined on average by the **linear equation**)<br>
> 
>    and there are additional assumptions; but, a deeper reflection on these is "beyond the scope" of STA130; nonetheless, they are that<br><br>
> - the $x_i$ **predictor variable** is **measured without error**
> - and the $\epsilon_i$ **errors** are **statistically independent** (so their values do not depend on each other)
> - and the $\epsilon_i$ **errors** are **unbiased** relative to the **expected value** of **outcome** $E[Y_i|x_i]=\beta_0 + \beta_1x_i$ (which is equivalently stated by saying that the mean of the **error distribution** is $0$, or again equivalently, that the **expected value** of the **errors** $E[\epsilon_i] = 0$)
